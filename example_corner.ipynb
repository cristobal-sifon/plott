Load modules and generate fake data

```python
%matplotlib inline
import pylab
from numpy import array, random

import plottools

random.seed(1938)

truth = [(11.6*(1+0.1*i), 25*(1+0.25*i)) for i in xrange(3)]
cov = array([[0.4,0.5],[0.5,5]])**2
chains = [random.multivariate_normal(xo, cov, 1000000).T for xo in truth]
```

Generate corner plot with several optionals (supressing warnings for clarity)

```python
figure = plottools.corner(chains,colors=('brown','blue','red'),
                          labels=[r'$p_1$', r'$p_2$'],
                          bins=[(20,25),(30,40),(30,40)],
                          clevels=(0.68,0.95),
                          ls2d=['-','--',':'],
                          ls1d=['-','--',':'],
                          medians1d=False,
                          percentiles1d=False,
                          style1d='curve',
                          background='filled',
                          bcolor=('orange','yellow'),
                          alpha=0.25,
                          names=['This Work','Previous Work','Earlier Work'],
                          verbose=False,
                          limits=[(10,16),(0,60)],
                          truths=truth[1],
                          truths_in_1d=True,
                          truth_color='black',
                          smooth=[1.,1.],
                          top_labels=False,
                          names_kwargs={'bbox_to_anchor': (0.55,0.85)})
```
